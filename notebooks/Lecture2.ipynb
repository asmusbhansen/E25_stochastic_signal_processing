### O2.1

**a)** Generate 100 samples of a discrete random variable with the following probability mass
function (PMF)

$
p_k=0.2, 0.6, 0.2
$
for
$
k=1,2,3
$

**b)** Estimate the PMF and cumulative distribution function (CDF) of the RV using those 100
samples and compare the resulting plots with the given theoretical PMF and CDF. Note
whether the estimated and theoretical functions match.
(Hint: Use relative frequency approach to estimate probabilities from finite samples of
RV)

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
vals = [1, 2, 3]
probs = [0.2, 0.6, 0.2]

# Sample from distribution
x = np.random.choice(vals, size=N, p=probs)

unique, counts = np.unique(x, return_counts=True)
counts_cumsum = np.cumsum(counts)
print(unique, counts/N)
print(unique, counts_cumsum/N)
```

### Problem O2.2

**a)** Plot the Binomial PMFs for $n=48$, $p=0.1, 0.3, 0.5, 0.75$

```python
from scipy.stats import binom

N = 48
probs =  [0.1, 0.3, 0.5, 0.75]

# Possible values of X (0 to n successes)
k = np.arange(0, N)

plt.figure()
for p in probs:
    pmf = binom.pmf(k, N, p)
    plt.plot(k, pmf, label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial distribution PMF")
```

**b) and c)** Plot the and𝑃[𝑋 <= 𝑥] 𝑃[𝑋 > 𝑥]

```python
plt.figure()
for p in probs:
    pmf = binom.pmf(k, N, p)
    plt.plot(k, np.cumsum(pmf), label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial distribution P[X<=x](CDF)")

plt.figure()
for p in probs:
    pmf = binom.pmf(k, N, p)
    plt.plot(k, 1-np.cumsum(pmf), label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial distribution P[X>x](CDF)")

```

**d)** Plot histograms of the 1000 realizations of the above RVs and show that they
approximate to the theoretical PMF

```python
plt.figure()
for p in probs:
    x = np.random.binomial(n=N, p=p, size=1000)
    plt.hist(x,bins=100, density=True, label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Density")
plt.title("Binomial distribution Density")
```

**e)** Find the mean and variance of the RV and compare with the theoretical values

The mean of a binomial distribution is given by

$$
\mathbb{E}[X]=np
$$

The variance i given by

$$
Var(X)=np(1-p)
$$


```python
for p in probs:
    x = np.random.binomial(n=N, p=p, size=1000)
    mean, var = np.mean(x), np.var(x)
    mean_theoretical = N*p
    var_theoretical = N*p*(1-p)
    print(f"p={p}, Estimated: Mean={mean:.3f} Var={var:.3f}, Theoretical: Mean={mean_theoretical:.3f} Var={var_theoretical:.3f}")
```

### Problem O2.3

**a)** Plot the PMFs and CDFs of geometric RVs for $p=0.1,0.5,0.9$

```python
from scipy.stats import geom

N = 48
probs =  [0.1, 0.5, 0.9]

# Possible values of X (0 to n trials before success)
k = np.arange(0, N)

plt.figure()
for p in probs:
    pmf = geom.pmf(k, p)
    plt.plot(k, pmf, label=f"p={p}")
plt.legend()
plt.xlabel("Number of trials before success")
plt.ylabel("Probability")
plt.title("Geometric distribution PMF")
```

**b) and c)**Plot the and𝑃[𝑋 <= 𝑥] 𝑃[𝑋 > 𝑥] and show that the sum of the PMF of the random variable = 1

```python
plt.figure()
for p in probs:
    pmf = geom.pmf(k, p)
    plt.plot(k, np.cumsum(pmf), label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Geometric distribution P[X<=x](CDF)")

plt.figure()
for p in probs:
    pmf = geom.pmf(k, p)
    plt.plot(k, 1-np.cumsum(pmf), label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Geometric distribution P[X>x](CDF)")
```

**d)** Plot histograms of the 1000 realizations of the above RVs and show that they
approximate to the theoretical PMFs

```python
plt.figure()
for p in probs:
    x = np.random.geometric(p=p, size=1000)
    plt.hist(x, density=True, label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Density")
plt.title("Geometric distribution Density")
```

**e)** Find the mean and variance of the RV and compare with the theoretical values

The mean of a geometric distribution is given by

$$
\mathbb{E}[X]=\frac{1}{p}
$$

The variance i given by

$$
Var(X)=\frac{(1-p)}{p^2}
$$

```python
for p in probs:
    x = np.random.geometric(p=p, size=1000)
    mean, var = np.mean(x), np.var(x)
    mean_theoretical = 1/p
    var_theoretical = ((1-p))/(p**2)
    print(f"p={p}, Estimated: Mean={mean:.3f} Var={var:.3f}, Theoretical: Mean={mean_theoretical:.3f} Var={var_theoretical:.3f}")
```
