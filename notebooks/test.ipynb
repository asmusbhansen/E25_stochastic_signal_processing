```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(100)*0.1
y = np.sin(x)

plt.figure()
plt.plot(x,y)
```

The **Discrete Fourier Transform (DFT)** of a sequence 
$x[n]$ of length $N$ is defined as:

$$
X[k] = \sum_{n=0}^{N-1} x[n] \, e^{-j \frac{2\pi}{N}kn}, 
\quad k = 0, 1, \dots, N-1
$$

And the **Inverse Discrete Fourier Transform (IDFT)** is:

$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] \, e^{j \frac{2\pi}{N}kn}, 
\quad n = 0, 1, \dots, N-1
$$
