## Lecture Notes

### Probability Mass Function(PMF)

The PMF defines the probability that the random variable takes some excat value

$$p_x(x)=P[X=x]$$

$p_x(x)$ is defined only for discrete values $x_1, x_2,..,x_n$

The PMF has the following important properties

All probabilities are positive: $p_x(x) \geq 0$

The sum of all probabilites is 1: $\sum_{n} p_x(x_n) = 1$

### Cumulative Distribution Function(CDF)

Defines the probability of the event $\{X \leq x\}$, meaning that $X$ takes on some value smaller than or equal to $x$.

$F_x(x)=P[X \leq x]$ for $-\infty \le x \le \infty$

Properties of the CDF

The CDF is always positive: $F_x(x) \geq 0$  
As x increases, we approach 1: $\lim_{x \to \infty} F_x(x)=1$  
The probability that $a \le x \leq b$ is: $F_x(b)-F_x(a)$

### Probability Density Function(PDF)

Definition:

$Density = \frac{Mass}{Volume}$

The PDF is the derivative of the CDF: $f_x(x)=\frac{dF_x(x)}{dx}$

Using integration, we can calculate the probability that the random variable $X$ falls in the interval $a$ to $b$.

$$
P[a \le X \leq b]=\int_a^b f_x(x)dx
$$


### Expected Value

The expected value is the center of gravity of the PDF.

Continous RV: $\mathbb{E}[X]=\int_{-\infty}^{\infty} f_x(t)tdt$  

Discrete RV: $\mathbb{E}[X]=\sum_{k} p_k(x_k)x_k$

Propeties:  
$\mathbb{E}[c] = c$  
$\mathbb{E}[aX] = a \, \mathbb{E}[X]$


### Variance

The variance measures variation around the mean value.

The variance of a random variable $X$ with mean $\mu = \mathbb{E}[X]$ is defined as
$$
\text{Var}(X) = \mathbb{E}\!\left[(X - \mu)^2\right]
$$

Equivalent formula:
$$
\text{Var}(X) = \mathbb{E}[X^2] - \left(\mathbb{E}[X]\right)^2
$$

Properties:
Variance of constant: $\text{Var}(c) = 0$  
Shift invariance: $\text{Var}(X + c) = \text{Var}(X)$  
Scaling: $\text{Var}(aX) = a^2 \, \text{Var}(X)$

### Problem O2.2

**a)** Plot the Binomial PMFs for $n=48$, $p=0.1, 0.3, 0.5, 0.75$

```python
from scipy.stats import binom

N = 48
probs =  [0.1, 0.3, 0.5, 0.75]

# Possible values of X (0 to n successes)
k = np.arange(0, N)

plt.figure()
for p in probs:
    pmf = binom.pmf(k, N, p)
    plt.plot(k, pmf, label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial distribution PMF")
```

**b) and c)** Plot the and[ <= ] [ > ]

```python
plt.figure()
for p in probs:
    pmf = binom.pmf(k, N, p)
    plt.plot(k, np.cumsum(pmf), label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial distribution P[X<=x](CDF)")

plt.figure()
for p in probs:
    pmf = binom.pmf(k, N, p)
    plt.plot(k, 1-np.cumsum(pmf), label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Binomial distribution P[X>x](CDF)")

```

**d)** Plot histograms of the 1000 realizations of the above RVs and show that they
approximate to the theoretical PMF

```python
plt.figure()
for p in probs:
    x = np.random.binomial(n=N, p=p, size=1000)
    plt.hist(x,bins=100, density=True, label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Density")
plt.title("Binomial distribution Density")
```

**e)** Find the mean and variance of the RV and compare with the theoretical values

The mean of a binomial distribution is given by

$$
\mathbb{E}[X]=np
$$

The variance i given by

$$
Var(X)=np(1-p)
$$


```python
for p in probs:
    x = np.random.binomial(n=N, p=p, size=1000)
    mean, var = np.mean(x), np.var(x)
    mean_theoretical = N*p
    var_theoretical = N*p*(1-p)
    print(f"p={p}, Estimated: Mean={mean:.3f} Var={var:.3f}, Theoretical: Mean={mean_theoretical:.3f} Var={var_theoretical:.3f}")
```

### Problem O2.3

**a)** Plot the PMFs and CDFs of geometric RVs for $p=0.1,0.5,0.9$

```python
from scipy.stats import geom

N = 48
probs =  [0.1, 0.5, 0.9]

# Possible values of X (0 to n trials before success)
k = np.arange(0, N)

plt.figure()
for p in probs:
    pmf = geom.pmf(k, p)
    plt.plot(k, pmf, label=f"p={p}")
plt.legend()
plt.xlabel("Number of trials before success")
plt.ylabel("Probability")
plt.title("Geometric distribution PMF")
```

**b) and c)**Plot the and[ <= ] [ > ] and show that the sum of the PMF of the random variable = 1

```python
plt.figure()
for p in probs:
    pmf = geom.pmf(k, p)
    plt.plot(k, np.cumsum(pmf), label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Geometric distribution P[X<=x](CDF)")

plt.figure()
for p in probs:
    pmf = geom.pmf(k, p)
    plt.plot(k, 1-np.cumsum(pmf), label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.title("Geometric distribution P[X>x](CDF)")
```

**d)** Plot histograms of the 1000 realizations of the above RVs and show that they
approximate to the theoretical PMFs

```python
plt.figure()
for p in probs:
    x = np.random.geometric(p=p, size=1000)
    plt.hist(x, density=True, label=f"p={p}")
plt.legend()
plt.xlabel("Number of successes")
plt.ylabel("Density")
plt.title("Geometric distribution Density")
```

**e)** Find the mean and variance of the RV and compare with the theoretical values

The mean of a geometric distribution is given by

$$
\mathbb{E}[X]=\frac{1}{p}
$$

The variance i given by

$$
Var(X)=\frac{(1-p)}{p^2}
$$

```python
for p in probs:
    x = np.random.geometric(p=p, size=1000)
    mean, var = np.mean(x), np.var(x)
    mean_theoretical = 1/p
    var_theoretical = ((1-p))/(p**2)
    print(f"p={p}, Estimated: Mean={mean:.3f} Var={var:.3f}, Theoretical: Mean={mean_theoretical:.3f} Var={var_theoretical:.3f}")
```

```python
unique_z, counts_z = np.unique(z, return_counts=True)

# If Z=X+Y, then the PMF of Z will be the discrete convolution of PMF for X and PMF for Y
k = np.arange(0, 20)
pmf_x = binom.pmf(k=k, n=8, p=0.5)
pmf_y = binom.pmf(k=k, n=4, p=0.5)
pmf_z = np.convolve(pmf_x, pmf_y)

print(pmf_x, pmf_y)

plt.figure()
plt.plot(unique_z, counts_z/N, label=f"Z Estimated")
plt.plot(k, pmf_z[:len(k)], label=f"Z Theoretical")
plt.title("Relative frequencies")
plt.xlabel("Number of successes")
plt.ylabel("Relative frequnecy")
plt.legend()
```

### 3.9

A coin is tossed n times. Let the random variable Y be the difference between the number of heads and the number of tails in the n tosses of a coin. Assume $P[heads]=p$

**a)** Describe the sample space of $S$.

Assuming $P[heads]=p$ and $p$ is the probability of a single toss results in heads is p, then n must be even.

The sample space is every possible combination of heads and tails for n tosses. We have $2^n$ possible combinations each with a probability of $\frac{1}{2^n}$ 

We define $Y=N_{Heads} - N_{Tails}$.
Since we assume that n i even $Y = \{ -n, -n+2,.., n-2, n \}$ because $Y$ jumps in steps of 2.


**b)** Find the probability of the event $\{Y=0\}$

If $Y=0$, then $N_{Heads}=N_{Tails}$, so for n toses, half of them should be successes $k=\frac{n}{2}$

$$
P(Y = 0) = \binom{n}{n/2}p^{(n/2)}(1-p)^{(n-n/2)}
$$

**c)**

For a given number of heads $X$

$$
P(Y = y) = \binom{n}{|n/2-y|}p^{(n/2)}(1-p)^{(n-n/2)}
$$
