### B

The theoretical autocovariance is expressed as

$$
R_x(k)=\sum_{n=-\infty}^\infty h_n h_{n-k}
$$

Where $h_n=e^{-2n}u_n$ and $u_n$ is the step function.

$$
R_x(k)=\sum_{n=-\infty}^\infty e^{-2n}u_n e^{-2(n-k)}u_{n-k}
$$

For $u_n>0$ we need $n>0$ and for $u_{n-k}>0$ we need $n \geq k$.

$$
R_x(k)=\sum_{n=k}^\infty e^{-2n}e^{-2(n-k)}=e^{2k} \sum_{n=k}^\infty (e^{-4})^n 
$$

We recognize the geometric series $\sum_{n=k}^\infty (e^{-4})^n=\frac{(e^{-4})^k}{1-e^{-4}}$ and the autocovariance becomes

$$
R_x(k)=e^{2k}\frac{(e^{-4})^k}{1-e^{-4}}=\frac{e^{-2k}}{1-e^{-4}}
$$

The system is z-transformed into

$$
X(z)=aX(z)z^{-1}+U(z)-bU(z)z^{-1}
$$

And rewritten in terms of the transfer function

$$
H(z)=\frac{X(z)}{U(z)}=\frac{(1-bz^{-1})}{(1-az^{-1})}
$$

Since the input $u_n$ is white noise with $\mu_x=0$, $\sigma_x^2=1$ which has a flat spectrum we evaluate the frequency response to get the PSD of the system.

$$
S_x(\omega)=|H(e^{j \omega})|^2\sigma_x^2
$$

```python
from scipy import signal

def gen_data(N):
    w = np.random.normal(0, 1, N)
    x = np.zeros_like(w)
    for i in range(1, N):
        x[i] = a * x[i-1] + w[i] - b*w[i-1]
    return w, x

N = 10000
fs = 1
omega = np.linspace(-np.pi, np.pi, N)

a_ = [0.9, 0.2]
b_ = [0.2, 0.9]

for a, b in zip(a_, b_):

    w, x = gen_data(N)
    
    # Compute Periodogram    
    X = np.fft.fftshift(np.fft.fft(x))
    freqs = np.fft.fftshift(np.fft.fftfreq(N, d=1/fs))
    
    Pxx = (1.0 / N) * (np.abs(X) ** 2)
    
    # Compute analytical PSD
    H = (1-b*np.exp(-1j*omega))/(1-a*np.exp(-1j*omega))
    
    plt.figure()
    #plt.plot(f_norm*2*np.pi, Pxx_den, label="Periodogram")
    plt.plot(2*np.pi*freqs, Pxx, label="Periodogram")
    plt.plot(omega, np.abs(H)**2, label="PSD")
    plt.legend()
    plt.yscale("log")
    plt.title(f"PSD - a={a}, b={b}")
```

The definition of the PSD is

$$
S_x(f)=\mathbb{F}[R_x(\tau)]
$$

```python
def acorr_to_periodogram(acorr):
    
    Pxx = np.fft.fftshift(np.fft.fft(acorr))
    freqs = np.fft.fftshift(np.fft.fftfreq(len(acorr), d=1/fs))
    return Pxx, 2*np.pi*freqs
    
```
