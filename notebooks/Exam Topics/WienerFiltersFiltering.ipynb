### Wiener Filtering

#### Model

We have a measured signal $x[n]$ corrupted by noise $w[n]$, where we want to estimate $\theta=s[n]$ based on $[x[0], x[1],..,x[n]]$

$$
x[n] = s[n]+w[n]
$$

Where the signal and noise is uncorrelated, leading to the autocorrelation matrix

$$
\mathbf{C}_{xx}=\mathbf{R}_{ss}+\mathbf{R}_{ww}
$$

We find the autocariance matrix

$$
\mathbf{C}_{\theta x} = \mathbb{E}(s[n][x[0],,.,x[n]]) 
$$

$$
\mathbf{C}_{\theta x} = \mathbb{E}({s[n]s[0]+s[n]w[0]),..,s[n]s[n]+s[n]w[n])})=[r_{ss}[n],..,r_{ss}[0]]=\mathbf{r'_{ss}}^T 
$$

We then get $a=(\mathbf{R_{ss}}+\mathbf{R_{ww}})^{-1}\mathbf{r_{ss}'}$ so the estimator  becomes

$$
\widehat{s}[n]= \mathbf{r_{ss}'}^T (\mathbf{R_{ss}}+\mathbf{R_{ww}})^{-1} \mathbf{x}
$$

To express this in terms of an LTI system, we flip $\mathbf{a}$ such that $\mathbf{h}=\mathbf{a'}$




#### Wiener-Hopf Equations

Restructuring the results gives 

$$
(\mathbf{R_{ss}}+\mathbf{R_{ww}}) \mathbf{h} = \mathbf{r_{ss}}^T
$$

Which are the Wiener-Hopf Equations

### Filtering in practice

When filtering, the number of samples to use for estimation is chosen as parameter **P**.

If $P=3$, then $\mathbf{C_{xx}}$ will be a 3x3 matrix, and both $\mathbf{h}$ and $\mathbf{r_{ss}}$ are 1x3 vectors.
