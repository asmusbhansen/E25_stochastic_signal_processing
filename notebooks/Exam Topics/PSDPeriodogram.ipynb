## Power Spectral Density(PSD) of Random Processes

The PSD Measure the average rate of change and an ensamble.

The autocorrelation can be shown as a rate of change of a random process
This measn that if \\(R(0)-R(\tau)\\) is small, then the probability of a large change \\(x(t+\tau)-x(t)\\) is small. 
$$
P[(x(t+\tau)-x(t)) < \epsilon]=P[(x(t+\tau)-x(t))^2 < \epsilon^2] \leq \frac{2(R_x(0)-R_x(\tau))}{\epsilon^2}
$$

This is rewritten using Chesbyshev's inequality \\(P(Y\geq a) \leq \frac{\mathbb{E}[Y]}{a}\\)

### Einstin-Wiener-Kinchin Theorem

The PSD of a WSS random process is the Fourier Transform of the autocorrelation function.

$$
S_x(f)=\mathcal{F}[x(t)] = \int_{-\infty}^{\infty} R_x(\tau) e^{-j 2 \pi f \tau} d\tau
$$

This is real-valued(\\(R_x(\tau)\\) is symmetric), even(\\(R_x(\tau)\\) is even) and non-negative.

### Cross Power Spectral Density

If two random processes X and Y are jointly WSS(Idenpendently WSS and Cross-correlation only depends on time difference) then their cross PSD is given as

$$
S_{X,Y}(f)=\mathcal{F}[X(t+\tau)Y(t)\] 
$$

This is complex values function, since the autocorrelation is not symmetric. The phase information describes the timelag between the different frequencies

### Einstein-Wiener-Kinchin Theorem for Discrete Random Processes

The PSD of a Discrete Random process is given by the Fourdier Transform of the autocorrelation.

$$
S_X(f) = \mathcal{F}[R_X(k)] \\
= \sum_{n=-\infty}^{\infty} R_X(n) e^{-j 2 \pi f n}
$$

It is real-values, even and non-negative. It is periodic with a period of \\(-1/2 \leq f \leq 1/2\\) or \\(-\pi \leq \omega \leq \pi\\)


### Example - PSD of white noise

If \\(X_n\\) is uncorrelated white noise, the autocorrelation is \\(R_x(n)=\sigma_x^2\\) for \\(n=0\\) and zero otherwise.

$$
S_X(f) = \sum_{n=-\infty}^{\infty} R_X(n) e^{-j 2 \pi f n}=\sigma_x^2
$$

Since \\(e^{-j 2 \pi f 0}=1\\) and \\((R_x(n)\\)) is 0 for all values other than \\(n=0\\).

### Example - PSD of MA process

Given the process \\(Y_n=X_n+\alpha X_{n-1}\\) where \\(X_n\\) is white noise with zero mean and variance \\(\sigma_x^2\\)

We calculate the autocorrelation

$$
\mathbb{E}[(X_n+\alpha X_{n-1})(X_{n+k}+\alpha X_{n-1+k})]=\mathbb{E}[X_nX_{n+k}]+\alpha\mathbb{E}[X_n X_{n-1+k}]+\alpha\mathbb{E}[X_{n-1} X_{n+k}]+\alpha^2\mathbb{E}[X_{n-1} X_{n-1+k}]
$$

$$
R_x(k) = \begin{cases} \sigma_x^2(1+\alpha^2) & \text{if } k = 0 \\ \alpha \sigma_x^2 & \text{if } k = 1 \\ 0 & \text{if } k > 1 \end{cases}
$$

This yields the Discrete Fourier Transform

$$
S_y(f)=\sigma_x^2(1+\alpha^2) + \alpha \sigma_x^2 e^{-j 2 \pi f} + \alpha \sigma_x^2 e^{j 2 \pi f} = \sigma_x^2 (1 + \alpha^2 + \alpha e^{-j 2 \pi f} + \alpha e^{j 2 \pi f})=\sigma_x^2 (1 + \alpha^2 + 2\alpha cos(2 \pi f))
$$

### Periodogram

By definition, the PSD needs the true autocorrelation function, which is rarely available when working with real signals.

Let \\([X_0,..,X_k]\\) be \\(k\\) observations of the WSS random process \\(X_n\\).

The Fourier Transform is

$$
\tilde{x}_k(f)=\sum_{m=0}^{k-1}X_me^{-j 2 \pi f m}
$$

The periodogram(PSD Estimate) is then

$$
p_k(f)=\frac{1}{k}|\tilde{x}_k(f)|^2
$$



#### Expected value of periodgram

$$
\mathbb{E}[p_k(f)]=\frac{1}{k}\mathbb{E}[\tilde{x}(f)\tilde{x}^*(f)]=\frac{1}{k}\sum_{m=0}^{k-1}\sum_{i=0}^{k-1}R_x(m-i)e^{-j 2 \pi f (m-i)}
$$

By setting \\(m'=m-i\\)

$$
\mathbb{E}[p_k(f)]=\frac{1}{k} \sum_{m'=-(k-1)}^{k-1} (k-|m'|)R_x(m')e^{-j 2 \pi f (m')}
$$

We can see that this is a biased estimate the will converge to the true value as \\(k\\) grow(Asymptotically biased)

#### Variance of periodogram

The variance does not go to zero as k grows, it is proportional to the square of \\(S_x(f)\\)(PSD)

$$
Var[\tilde{p}_k(f)] = S_X^2(f) \left( 1 + \left( \frac{\sin(2\pi f k)}{k \sin(2\pi f)} \right)^2 \right)
$$

This is valis for general signals which are not WGN.


#### Smoothing Periodograms

We can smooth the periodograms to decrease the variance.

$$
\left< \tilde{p}_k(f) \right>_N = \frac{1}{N} \sum_{i=1}^{N} \tilde{p}_{k,i}(f)
$$

Where \\(\tilde{p}_{k,i}(f)\\) is the i'th periodogram of N each created using k samples.

#### Mean of smoothed periodogram

$$
E[\left< \tilde{p}_k(f) \right>_N] = \frac{1}{N} \sum_{i=1}^{N} E[\tilde{p}_{k,i}(f)] = E[\tilde{p}_k(f)]
$$

#### Variance of smoothed periodogram

$$
Var[\left< \tilde{p}_k(f) \right>_N] = \frac{1}{N^2} \sum_{i=1}^{N} \mathrm{VAR}[\tilde{p}_{k,i}(f)] = \frac{1}{N} \mathrm{VAR}[\tilde{p}_k(f)]
$$

Since the \\(Var(N x)=N^2Var(x)\\) and the periodograms are independent

#### Bartletts Smoothing Procedure

 1. \\(Nk\\) samples are divided into N partitions of k samples
 2. The periodogram is computed for each partition
 3. The smoothed periodogram is computed by averaging each freqeucny in over all N periodograms

The k parameters determines the frequency resolution: \\(\Delta f = \frac{f_s}{N}\\)
Greater k, better frequency resolution.

The N parameters determines the variance. Greater N, lower variance.
