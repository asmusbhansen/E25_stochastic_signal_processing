#### Model

The vocal tract is modelled as a variable resonater(All-pole filter) where either a pulse train with a certain pitch or random noise is used for excitation.



#### All-Pole Filter

The system equation is given as $X(z)=H(z)E(z)$ where a signal excites the all-pole filter

In an all pole filter, we have no roots in the numerator, and roots in the denominator.

$$
H(z)=\frac{1}{1-\sum_{k=1}^p a_k z^{-k}}
$$

We can rewrite the system equation as 

$$
X(z)(1-\sum_{k=1}^p a_k z^{-k})=E(z)
$$

Then the inverse z transform is performed, using the fact that $X(z)z^{-k}$ becomes $x(n-k)$.

$$
x(n)-\sum_{k=1}^p a_k x(n-k)=e(n)
$$

$$
x(n)=\sum_{k=1}^p a_k x(n-k+e(n)
$$

This is similar to an autoregressive model where the current sample depends on a conbination of past samples and an error/excitation term.



#### Least Square solution to LPC

For a p term all pole filter, we have the equations

$$
\mathbf{x_i}=[x[n-1+i], x[n-2+i],..,x[n-p+i]]
$$

$$
\mathbf{b}=[x(n), x[n+1],..,x[n+N]]
$$

$$
\mathbf{e}=[e(n), e[n+1],..,e[n+N]]
$$

$$
\mathbf{A}=[\mathbf{x_0}, \mathbf{x_1},..,\mathbf{x_{n+N}}]
$$

Where **b** and **e** are Nx1 vectors, **a** is a px1 vector and **A** is a Nxp matrix.

We try to find coeffecients $\mathbf{a}$ to minimize the error term

$$
\mathbf{e}=\mathbf{b}-\mathbf{A}\mathbf{a}
$$

This can be done using least squares where

$$
\mathbf{a}=(\mathbf{A}^T\mathbf{A})^{-1}\mathbf{A}^T\mathbf{b}
$$



### Maximum Likelihood Solution to LPC

The maximum likelihood solution to LPC is the Yule-Walker equations

$$
\mathbf{R_x}\mathbf{a}=\mathbf{r_x}
$$

Where $\mathbf{R_x}$ is the symmetric Toeplitz autocorrelation matrix with values $R_x(0)$ to $R_x(p-1)$ and is $\mathbf{r_x}=[R_x(1), R_x(2),..,R_x(p)]$

We can show the the least squares solution is the same as the maximum likelihood solution.

Least squares: $\mathbf{A}\mathbf{a}+\mathbf{e}=\mathbf{b}$

If we multiply by $\mathbf{A}^T$ and take the expectation on both sides

$$
\mathbb{E}[\mathbf{A}^T\mathbf{A}]\mathbf{a}+\mathbb{E}[\mathbf{A}^T\mathbf{e}]=\mathbb{E}[\mathbf{A}^T\mathbf{b}]
$$

Then $\mathbb{E}[\mathbf{A}^T\mathbf{A}]=R_x$, $\mathbb{E}[\mathbf{A}^T\mathbf{e}]=0$ and $\mathbb{E}[\mathbf{A}^T\mathbf{e}]=[R_x(1), R_x(2),..,R_x(p)]$ since $\mathbf{A}$ is offset one sample from $\mathbf{b}$ and $\mathbf{e}$.


### Linear Predictive Coding in Practice

These are the steps used in LPC

#### Encoding

1. Divide signal into blocks
2. (Optional) Apply windowing to reduce discontinuties in the edges
3. Determine voiced/unvoiced speech and pitch
4. Determine Filter coeffecients $\mathbf{a}$ and exctation $\mathbf{e}$

#### Decoding

1. Pass exication $\mathbf{e}$ though filter coeffecients $\mathbf{a}$
2. Inverse step 2 in encoding, but assembling signal. Depends on overlaps and window functions
