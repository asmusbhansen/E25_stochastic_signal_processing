###  Wiener Prediction

#### Objective

Estimating $\theta=x[N-1+I]$ for $I \geq 1$ based on $[x[0],..x[N-1]]$. This is an $I$ step ahead predictor.

The estimator becomes

$$
\widehat{x}[N-1+I]=\mathbf{r_{xx}'}^T\mathbf{R_{xx}}^{-1}\mathbf{x}=\mathbf{a}^T\mathbf{x}
$$

We define the cross-covariance matrix

$$
\mathbf{C_{\theta x}} = \mathbb{E}[x[N-1+I][x[0],..x[N-1]]]
$$

$$
\mathbf{C_{\theta x}} = \mathbb{E}[x[N-1+I]x[0],..x[N-1+I]x[N-1]]=[r_{xx}[N-1+I],..,r_{xx}[I]]=\mathbf{r_{xx}'}^T
$$

We then get $\mathbf{a}=\mathbf{R_{xx}}^{-1}\mathbf{r_{xx}'}$

To express this in terms of an LTI system, we flip $\mathbf{a}$ such that $\mathbf{h}=\mathbf{a'}$

#### Wiener-Hopf Equations

Rewriting $\mathbf{a}=\mathbf{R_{xx}}^{-1}\mathbf{r_{xx}'}$ results gives 

$$
\mathbf{R_{xx}}\mathbf{h} = \mathbf{r_{ss}}^T
$$

Which are the Wiener-Hopf Equations

### Prediction in practice

When prediction, the number of samples to use for prediction is chosen as parameter **P**.

If $P=3$, then $\mathbf{R_{xx}}$ will be a 3x3 matrix, and both $\mathbf{h}$ and $\mathbf{r_{xx}}$ are 1x3 vectors.
