### Stationarity

#### Random Process

Is a model for a signal which has a non-deterministic behaviour and is governed by probability

Example - $X(t)$ is the coin toss a time $t$ which has value $[0, 1]$

#### Realization

Single realizaton of $k$ coin tosses. Will have one of many possible outcomes for the random process.

#### Discrete Vs. Continous Random Process

If $t$ is discrete, then the random process is discrete, otherwise continous.




#### Description of a Random Process

The randon process can be described by the CDF

$$
F_{x_1x_2,..,x_k}(x_1,x_2,..,x_k)=P[X_1 \leq x_1,X_2 \leq x_2,..,X_k \leq x_k]
$$

The Random Process i a collection of k'th order joint CDF's. We would need a large number of CDF's to describe the random process at all values og k, to all time indexes t.

##### Continous Vs. Discrete Random Describtion

For continous, the describe the random process by it's PDF, $f_{x_1,x_2,..,x_k}(x_1,x_2,..,x_k)$

For discrete, is described by the PMF $p_{x_1,x_2,..,x_k}(x_1,x_2,..,x_k)=P[X_1=x_1,X_2=x_2,..,X_k=x_k]$


#### Strict Stationarity

Any set of samples are identically distributed to the shifted set

$$
F_{x_1x_2,..,x_k}(x_1,x_2,..,x_k)=F_{x_{1+\tau},x_{k+\tau}}(x_{1+\tau},..,x_{k+\tau})
$$

#### Wide Sense Stationary(WSS)

The strict sense stationarity it not practical, since we don't have a complete discription of out system when working in the real world. Instead we use a less restrictive stationarity called WSS.

For a process to be WSS, it has to have **constant mean** and only **lag-dependent auto covariance/correlation**

All strict sense stationary processes are WSS but not inversely.

#### Estimation of mean and autocorrelation

WSS allows us to estimate mean and autocorrelation using a single realization.
This property is called **Ergodicity**, meaning that time averages are equal to ensample averaging.

$$
\bar{x}=\frac{1}{N}\sum_{i=1}^N x_i
$$

$$
\gamma_x(k)=\frac{1}{N}\sum_{i=1}^{N-h} (x_{i+h}-\bar{x})(x_i-\bar{x})
$$


#### Power Spectral Density

Definition $S_x(f)=\mathcal{F}(R_x(k))$

For the PSD to be well defined, we need a process that is WSS such that the autocorrelation only depends on the sample/time lag k.

#### Linear Time-Invariant Systems

If a WSS process is passed though a LTI system $y_t=h_t \ast x_t$ then the output in the frequency domain is.

$$
S_y(f)=|H(f)|^2 S_x(f)
$$


#### Gaussian Process as special case

Consider the gaussian process $[X_1, X_2,..,X_k]$ described by a multivariate Gaussian.

$$
\mathbf{X} \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\mu})
$$

Since this process is completely described by the mean vector $\boldsymbol{\mu}$ and covariance $\boldsymbol{\mu}$, it is strictly stationary if it is also WSS.

To be WSS, the mean has to be constant, and the covariance matrix has to be Toeplitz and symmetric, since then the covariance will only depend on the time/sample lag k.

Single variate 
$$
f(x)=\frac{1}{\sqrt{2\pi}\sigma}e^{-\frac{(x-\mu_x)^2}{2\sigma^2}}
$$

Multi variate
$$
f_{\mathbf{X}}(\mathbf{x}) = \frac{1}{\sqrt{(2\pi)^D |\boldsymbol{\Sigma}|}} \exp \left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{x} - \boldsymbol{\mu}) \right)
$$



### Models

#### White Noise

Let $X_t \sim N(0, \sigma_w^2)$, then all samples are random and independently distributed.



#### Moving Average(MA)

Consider the MA(1) model. Let $y_t=\frac{1}{2}(w_t+w_{t-1})$

##### Mean of MA

$\mathbb{E}[y_t]=\mathbb{E}[\frac{1}{2}(w_t+w_{t-1})]=0$ since $\mathbb{E}[w_t]=0$ 

##### Autocorrelation of MA

$\mathbb{E}[y_ty_{t+k}]=\mathbb{E}[\frac{1}{2}(w_{t}+w_{t-1})\frac{1}{2}(w_{t+k}+w_{t-1+k})]=\frac{1}{4}(\mathbb{E}[w_{t}w_{t+k}]+\mathbb{E}[w_{t}w_{t-1+k}]+\mathbb{E}[w_{t-1}w_{t+k}]+\mathbb{E}[w_{t-1}w_{t-1+k}])$

For $k=0$, this becomes $\mathbb{E}[y_ty_{t+k}]=\frac{1}{2}\sigma_w^2$

For $|k| = 1$, $\mathbb{E}[y_ty_{t+k}]=\frac{1}{4}\sigma_w^2$

##### Stationarity

The model is WSS since mean is constant and autocorrelation is only dependent on the time lag.

#### Autoregressive(AR)

Consider the AR(1) model. Let $y_t=\alpha y_{t-1}+w_t$

##### Mean of AR

$\mathbb{E}[y_t]=\mathbb{E}[\alpha y_{t-1}+w_t]=\alpha\mathbb{E}[y_{t-1}]$

This is recursive, and if $|\alpha| < 1$, then the mean will converge towards 0 regardless of inital value $X_0$

##### Autocorrelation of AR

We're multiplying by $y_{t+k}$ on both sides and taking the expectation

$\mathbb{E}[y_t y_{t+k}]= \alpha\mathbb{E}[ y_{t-1} y_{t+k}] + \mathbb{E}[w_t y_{t+k}]$

Since $\mathbb{E}[ y_{t-1} y_{t+k}]=R_y(k)$

$R_y(k)= \alpha R_y(k-1)+\mathbb{E}[w_t y_{t+k}]$

For $|k|>0$, $R_y(k)= \alpha R_y(k-1)$ we can rewrite this recursion to $R_y(k)=a^kR_y(0)$ 

For $k=0$, $R_y(0)= \alpha R_y(1)+\sigma_w^2$

Since past values $w_{t-k}$ is not correlated to current.

$\mathbb{E}[w_t y_t]=\mathbb{E}[w_t (\alpha y_{t-1}+w_t)]=\mathbb{E}[w_t w_t]=\sigma_w^2$

Then we can find the variance $R_y(0)=\alpha R_y(1)+\sigma_w^2=\alpha^2 R_y(0)+\sigma_w^2$

Then $R_y(0)(1-\alpha)=\sigma_w^2$ is rewritten to $R_y(0)=\frac{\sigma_w^2}{(1-\alpha)}$

##### Stationarity

The mean is constant and the autocorrelation only depends on the lag k, so the process is WSS.

##### Randow Walk with dritft

Let $x_t=\delta+x_{t-1}+w_t$ where $\delta$ is drift.

This can also be written as $x_t=\delta t + \sum_{i=0}^t w_i$

##### Mean 

$\mathbb{E}[x_t]=\mathbb{E}[\delta+x_{t-1}+w_t]=\delta + \mathbb{E}[x_{t-1}]$

This is recursive, and can be rewritten to $\mathbb{E}[x_t]=\delta t$

##### Autocorrelation

For this we are setting $\delta=0$. 

$\mathbb{E}[x_t x_s]=\mathbb{E}[\sum_{i=0}^t w_i \sum_{j=0}^d w_j]$

For $t \geq s$ we will have $s$ times where $i=j$ 

$\mathbb{E}[x_t x_s]=\sigma_w^2 s$

For $s \geq t$ then $\mathbb{E}[x_t x_s]=\sigma_w^2 t$

Hence, $\mathbb{E}[x_t x_s]=\sigma_w^2 min(s,t)$

##### Stationarity

We can see that this process is **NOT** stationary, since the mean depends on time, and the autocorrelation depends on time $s$ and $t$ and not $s-t$
